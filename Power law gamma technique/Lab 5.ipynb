```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('apples.jpg', 1)

# Convert image into RGB image
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
plt.show()

# Convert color image to Negative Transformation
img_neg = cv2.bitwise_not(img_RGB)

plt.imshow(img_neg)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('gamma.jpg', 0)

plt.imshow(img1, cmap = 'gray')
plt.show()

# Calculate the power law / gamma transformation
gamma = 0.6
img_power1 = np.power(img1, gamma)

plt.imshow(img_power1, cmap = 'gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img2 = cv2.imread('log.jpg', 0)

plt.imshow(img2, cmap = 'gray')
plt.show()

# Calculate log transformation 
np.seterr(divide = 'ignore')
c = 255 / np.log(1 + np.max(img2))
# c = 255 / np.log(1 +255)
log_image = c * (np.log(img2 + 1))
   
# Specify the data type so that
# float value will be converted to int
log_image = np.array(log_image, dtype = np.uint8)

plt.imshow(log_image, cmap = 'gray')
plt.show()
```
